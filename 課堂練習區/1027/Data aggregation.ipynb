```python
import pandas as pd
data = pd.read_csv("bicycle_road.csv",index_col=0)
data.head()
```

```python
print("資料(資料筆數,欄位數量): ",data.shape)
print("資料中所有欄位名稱: ",data.columns)
print("資料中是否有空值: ",data.empty)
```

```python
set(data["縣市"])
```

```python
c1 = (data["縣市"] == "高雄市")
d1 = data[c1]

print("共有",len(d1),"筆資料")
d1.head()
```

```python
#篩選資料:多條件
c2 = (data["縣市"] == "高雄市") & (data["挑戰指數"].str.contains("2級"))
d2 = data[c2]

print("共有",len(d2),"筆資料")
d2.head()
```

```python
#資料名稱.loc[行標籤,列標籤]
loc1 = data.loc[10:15,["自行車道路名稱","簡介","總計長度"]]
loc1
```

```python
#資料名稱.iloc[行索引,列索引]
iloc1 = data.iloc[20:26,0:3]
iloc1
```
