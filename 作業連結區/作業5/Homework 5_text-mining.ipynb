```python
import jieba
import jieba.analyse
import pandas as pd
import numpy
from collections import Counter
```

```python
orgdata = pd.read_csv('shops_zh-tw.csv')
```

```python
documents = ['我是台灣師範大學科技系柯幸孜']
# 精確模式
for sentence in documents:
    seg_list = jieba.cut(sentence)
    print('/'.join(seg_list))
```

```python
jieba.load_userdict('./userdict.txt')

for sentence in documents:
    seg_list = jieba.cut(sentence)
    print('/'.join(seg_list))
```

```python
orgdata['seg'] = ""
orgdata.head()
```

```python
jieba.load_userdict('./userdict.txt')
jieba.analyse.set_stop_words('stopwords.txt')
index = 0
for documents in orgdata['introduction'].astype(str):
    seg_list = jieba.analyse.extract_tags(documents, 150)
    #seg_list = jieba.lcut(documents)
    orgdata['seg'].iloc[index] = seg_list
    index += 1
```

```python
orgdata['seg']
```

```python
date_split = orgdata['category'].iloc[0].split('，')
```

```python
date_split[0]
```

```python
orgdata['cate'] = ""
index = 0
for documents in orgdata['category'].astype(str):
    date_split = documents.split('、')
    orgdata['cate'].iloc[index] = date_split[0]
    index += 1
```

```python
orgdata.to_csv("shop.csv", index=False)
```
