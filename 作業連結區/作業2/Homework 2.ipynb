### 查詢各項目的不同資料排序

```python
#定義datasorted()函數，得到不同細項的資料排序
def datasorted(s):
    
    allcategory = []
    allcategory = temp.get((s))
    allcategory

    alldata = []
    x = input("輸入資料:")
    for i in allcategory:
        alldata.append(i.get((x)))

    alldata_enumerate = enumerate(alldata, start = 1)
    alldata_enumerate_sorted = sorted(alldata_enumerate, key = lambda f: float(f[1]), reverse = True) #使用float將浮點數排序
    print(s,"的",x,"排序是")
    for l in alldata_enumerate_sorted:
        j = l[0] - 1
        print(allcategory[j].get('細項'), allcategory[j].get(x))
```

```python
y = input("輸入項目別:")
datasorted(y)
```
