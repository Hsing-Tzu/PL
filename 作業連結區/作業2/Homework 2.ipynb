### 定義gender()函數，儲存性別內容

```python
def gender(dict_滿意度調查): 
    number1 = len(dict_滿意度調查['性別']) 
    #len()函式用來計算dict性別共有幾個key
    content1 = []
    for i in dict_滿意度調查['性別']: #for detail1 in dict_name['性別'].items:
        detail1 = (i,dict_滿意度調查['性別'][i]) #用 .items 這行就可以不用
        content1.append(detail1)
    return number1, content1
```

```python
for n in data.items():
    a1, b1 = gender(n[1])
    print(a1)
    print(b1)
```
